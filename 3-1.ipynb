결측치(null값) 확인하기 - data 전처리

```python
# null 값의 평균을 계산하여 컬럼별 null의 비율 확인
df.isnull().mean().sort_values()
```

## categorical column 확인하기

```python
cat_columns = [col for col in df.columns if df [col].dtypes == 'object']
print(cat_columns)
```

## numerical column 확인하기

```python
# numerical column 목록만 추출

num_columns = [num for num in df.columns if df [num].dtypes == 'float64']
print(num_columns)

# dtype이 float64 타입인 것 추출하는 것 말고
# object 반대 값이 숫자 값이므로 != object해도 됨
```

## 결측치 개수 확인하기

```python
# categorical column의 null 비욜(null이 있는 지 확인)
# df[cat_columns].isnull()
df[cat_columns].isnull().mean().sort_values()
```

```python
# numerical column의 null 비욜
df[num_columns].isnull().mean().sort_values()
```

```python
# numerical 컬럼에서 null 값을 중위값으로 결측치 채우기
for col in num_columns:
    if df[col].isnull().mean() > 0: # 컬럼에 널이 있는 경우
        col_median = df[col].median() # 컬럼의 중앙 값 계산
        df[col].fillna(col_median, inplace = True)  # 결측치 채우기
```

```python
# categorical 컬럼에서 null 값의 갯 수 평균이 0이상인 것 출력 확인
for col in cat_columns:
    if df[col].isnull().mean() > 0:
        print(col, (df[col].isnull().mean()))
```

```python
# categorical value를 가진 column은 최빈값(mode)으로 결측치 채우기
df['WindGustDir'].fillna(df['WindGustDir'].mode()[0], inplace=True)
df['WindDir9am'].fillna(df['WindDir9am'].mode()[0], inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0], inplace=True)
df['RainToday'].fillna(df['RainToday'].mode()[0], inplace=True)
```

```python
# 날짜는 필요하지 않음으로 column drop
df.drop(['Date'], axis = 1, inplace = True)
```

결정 트리(Decision Tree)로 내일 비가 올지 안올지 예측해보기
